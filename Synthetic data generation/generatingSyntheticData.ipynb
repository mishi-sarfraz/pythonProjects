```python
import pymongo
import pandas as pd
# db_server=mongodb address of shopgram data 

db_server = "mongodb://mongoadmin:"
 
def save_to_db(df,db_server,db_name, collection, del_db = False):
	"""
	Saves a dataframe to a mongoDB collection
	Delets all the data in the collection if del_db = True
	"""
	client = pymongo.MongoClient(db_server)
	#for testing
	#importing database
	db  = client[db_name]
	if del_db != False:
		if collection in db.list_collection_names():
			# if db.collection_names().count('asiaOne_keywords') > 0:
			db[collection].delete_many({})
			
	data = df.to_dict(orient="records")
	db[collection].insert_many(data)
	client.close()


def read_from_db_by_date(db_server,db_name,collection,start_date,end_date):
	"""
	Reads from a mongoDB collection by date
    Parameters
    ----------
    db_server : str
        The mongoDB server to connect to
    collection : str    
        The collection to read from
    start_date : datetime    
        The start date to read from
    end_date : datetime  
        The end date to read from
    Returns
    -------
    df : dataframe
        The dataframe read from the collection
	"""
	client = pymongo.MongoClient(db_server)
	db  = client[db_name]
	#find the data filtered by the ga:date column with start date and end date in python 3
	#convert the dates to unix time with 00 hour 00 minute 00 second and 000000 microsecond
	start_date = int(start_date.timestamp()) *1000000000
	end_date = int(end_date.timestamp()) *1000000000
	#find the entries in teh mongo db that are between the start and end date
	data = db[collection].find({'ga:date':{'$gte':start_date,'$lte':end_date}})
	df = pd.DataFrame(list(data))
	#convert ga:date in df to datetime
	df['ga:date'] = pd.to_datetime(df['ga:date'])
	client.close()

	return df


def db_get_articles_by_url(db_server,db_name,collection,df):
    """
    This function takes a dataframe and returns a dataframe with the articles from the mongodb
    
    """
    client = pymongo.MongoClient(db_server)
    db  = client[db_name]
    #find the articles based on the route column of df 
    data = db[collection].find({'path':{'$in':df['path'].tolist()}})

    df = pd.DataFrame(list(data))

    #convert ga:date in df to datetime

    client.close()
    return df

#read all rows from the collection
def read_from_db(db_server,db_name, collection):
    """
    Reads from a mongoDB collection
    Parameters
    ----------
    db_server : str
        The mongoDB server to connect to
    collection : str
        The collection to read from
    Returns
    ------- 
    df : dataframe
        The dataframe read from the collection
    """
    client = pymongo.MongoClient(db_server)
    db  = client[db_name]
    data = db[collection].find()
    df = pd.DataFrame(list(data))
    client.close()
    return df
    
def db_check_not_empty(db_server,db_name,collection):
    """
    Checks if a collection is empty
    Parameters
    ----------
    db_server : str
        The mongoDB server to connect to
    collection : str    
        The collection to check
    Returns 
    ------- 
    """
    client = pymongo.MongoClient(db_server)
    db  = client[db_name]
    #data = db[collection].find()
    #if statement to check if the collection is empty, use count_documents to check if the collection is empty
    if db[collection].count_documents({}) == 0:
        return False
    else:
        return True

def db_get_newest_entry(db_server,db_name,collection):
    """
    Gets the newest entry in a collection
    Parameters
    ----------
    db_server : str
        The mongoDB server to connect to
    collection : str
        The collection to check
    Returns
    -------
    df : dataframe
        The newest entry in the collection
    """
    client = pymongo.MongoClient(db_server)
    db  = client[db_name]
    data = db[collection].find().sort('_id',-1).limit(1)
    df = pd.DataFrame(list(data))
    client.close()
    return df
```
