#### Dataframe 2 EIBTDateTimeOffset

```python
# Convert df2_cleaned_updated['EIBTDateTimeOffset'] to datetime
df2_cleaned_updated['EIBTDateTimeOffset'] = pd.to_datetime(df2_cleaned_updated['EIBTDateTimeOffset'])
df2_cleaned_updated['EIBTDateTimeOffset'].head(3)
```

```python
df2_cleaned_updated.info()
```

```python
# remove offset in EIBTDatetimeOffset 
df2_cleaned_updated['EIBTDateTimeOffset'] = df2_cleaned_updated['EIBTDateTimeOffset'].dt.tz_localize(None)
df2_cleaned_updated.head(3)
```

```python
# check df1_latest_update['EIBT'] and df2_cleaned_updated['EIBTDateTimeOffset'] are in different format
# df1_latest_update['ArrivalFlight'] and df2_cleaned_updated['ArrivalFlight'] are same
```

#### Merging on key columns

```python
print(df1_latest_update.columns)
df2_cleaned_updated.columns
```

```python
print(df1_latest_update.info())
df2_cleaned_updated.info()
```

```python
# check duplicate rows
print(df1_latest_update.duplicated().sum())
df2_cleaned_updated.duplicated().sum()
```

```python
# Merge df1_latest_update and df2_cleaned_updated
df_merged = pd.merge(df1_latest_update,df2_cleaned_updated,how='inner',left_on=['ArrivalFlight','EIBT','DepartureFlight'],
                                                        right_on=['ArrivalFlight','EIBTDateTimeOffset','DepartureFlight'])
df_merged
```

```python
# null values in df_merged
df_merged.isnull().sum().sort_values(ascending=False)
```

```python
# check duplicate rows
df_merged.duplicated().sum()
```

```python
# round off success rate to 2 decimal places
df_merged['% Success'] = df_merged['% Success'].round(2)
df_merged
```

```python
# keep only those rows where % Success is equal to 100
df_merged = df_merged[df_merged['% Success'] == 1.00]
df_merged
```

```python
# value counts % Success
df_merged['% Success'].value_counts()
```

```python
# to csv file
df_merged.to_csv('Merged_Flight_Pairs.csv',index=False)
```
